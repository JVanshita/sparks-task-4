```python
terror = terror[pd.notnull(terror.lat)]
terror = terror[pd.notnull(terror.lon)]
print("unknown values of latitude and longitude are removed")
```

```python
exclude_cols=['year','lat','lon']
float_cols = [c for c in terror.select_dtypes(include=[float]).columns.tolist() if c not in exclude_cols]
```

```python
terror[float_cols]= terror[float_cols].fillna(0).astype(int)
```

```python
terror[float_cols]= terror[float_cols].mask(terror[float_cols]<0,0)
```

```python
terror['wounded'] = terror['wounded'].fillna(0)
terror['killed'] = terror['killed'].fillna(0)
```

```python
terror['casualties']= terror['killed']+terror['wounded']
```

```python
terror['city']= terror['city'].replace(['unknown'],'bagdad')
```

```python
# 1. Distribution of terrorist attacks over regions from 1970-2017
```

```python
chunk_size = 10000  # You can increase or decrease this based on available memory

# Initialize an empty DataFrame to store the crosstab results
final_crosstab = None

# Create an iterator to read the CSV file in chunks
csv_iterator = pd.read_csv("D:\data analysis\globalterrorismdb_0718dist.csv", encoding='latin-1', chunksize=chunk_size)

# Iterate over chunks and aggregate data
for chunk in csv_iterator:
    # Assuming your dataset has 'year' and 'region' columns
    crosstab_chunk = pd.crosstab(chunk['iyear'], chunk['region_txt'])
    
    # Combine the chunk's crosstab with the final result
    if final_crosstab is None:
        final_crosstab = crosstab_chunk
    else:
        final_crosstab = final_crosstab.add(crosstab_chunk, fill_value=0)

# Now you have the final crosstabulation in 'final_crosstab' variable
final_crosstab.head()
```

```python
final_crosstab.plot(kind="area",stacked=False, alpha=0.5 ,figsize=(20,10))
plt.title("Region wise Attacks", fontsize=15)
plt.xlabel("Years",fontsize=15)
plt.ylabel("Number of Attacks", fontsize=15)
plt.show()
```

```python
regt=final_crosstab.transpose()
regt["total"]=regt.sum(axis=1)
ra=regt["total"].sort_values(ascending=False)
ra
```

```python
ra.plot(kind="bar", figsize=(15,6))
plt.title("Total Number Of Attacks In each Region from 1970-2017", fontsize=15)
plt.xlabel("Regions",fontsize=15)
plt.ylabel("Number of Attacks", fontsize=15)
plt.show()
```
